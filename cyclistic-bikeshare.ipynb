# Changing the datatype "object" to "datetime"

```python
file["started_at"]= pd.to_datetime(file["started_at"])
file["ended_at"]= pd.to_datetime(file["ended_at"])
file.info()
```

# Removing duplicate columns

```python
clean_file=file.dropna()
clean_file.info()
```

```python
clean_file.isnull()
```

# calculating the riding time by (ended_at - started_at)

```python
clean_file["ride_length"]= clean_file["ended_at"] - clean_file["started_at"]
clean_file["ride_length"]
```

# Extracting  only date from started_at and adding new column as started_date

```python
clean_file["started_date"]=[i.date()for i in clean_file["started_at"]]
clean_file["started_date"]

```

# converting started_date datatype into datetime

```python
clean_file["started_date"] = pd.to_datetime(clean_file["started_date"])
clean_file.info()
```

# finding weekdays from started_date and add a new column for weekday values

```python
#finding weekdays from date
clean_file ["week_day"] = clean_file["started_date"].apply(lambda x:x.weekday())
clean_file["week_day"].unique()
```

```python
#converting into weekdays
dict={0:"sunday",1:"monday",2:"tuesday",3:"wednessday",4:"thursday",5:"friday",6:"saturday"}
clean_file["week_day"]= clean_file["week_day"].apply(lambda y: dict[y])
clean_file["week_day"]
```

# Extracting year and month from started_date

```python
clean_file["year"]= pd.DatetimeIndex(clean_file["started_date"]).year
clean_file["month"]= pd.DatetimeIndex(clean_file["started_date"]).month
clean_file.head(10)
```

# Calculating the distance between starting and ending point

```python
clean_file["x"]=(clean_file["end_lat"] - clean_file["start_lat"])**2
clean_file["y"]=(clean_file["end_lng"] - clean_file["start_lng"])**2
clean_file["distance"]= (clean_file["x"]+clean_file["y"])**(1/2)
clean_file["distance"]
```

# Sort data in asc order by date

```python
clean_file.sort_values(by=["started_at"],inplace =True, ascending = True)
clean_file.head(5)
```

# Calculate the number of rides for all types of users, annual members and casual riders by weekday by adding count of ride_id to values.

```python
#calculate number of rides for all types of users by weekday by adding count of ride_id to values
count_users_weekday = clean_file.groupby("week_day")["ride_id"].count()
count_users_weekday = count_users_weekday.sort_index()
count_users_weekday
```

```python
#calculate number of rides for members by weekday by adding count of ride_id to values
count_weekday_member = clean_file[clean_file["member_casual"]=="member"].groupby("week_day")["ride_id"].count()
count_weekday_member = count_weekday_member.sort_index()
count_weekday_member
```

```python
#calculate number of rides for casual by weekday by adding count of ride_id to values
count_weekday_casual = clean_file[clean_file["member_casual"]=="casual"].groupby("week_day")["ride_id"].count()
count_weekday_casual = count_weekday_casual.sort_index()
count_weekday_casual
```

# To plot weekday vs number of rides

```python
plt.figure(figsize =(10,8))
plt.plot(count_users_weekday.index,count_users_weekday.values)
plt.plot(count_weekday_member.index,count_weekday_member.values)
plt.plot(count_weekday_casual.index,count_weekday_casual.values)
plt.title("Calculating the number of rides for different users by weekday by adding count of ride_id to values")
plt.legend(["user","member","casual"])
labels = ["Sunday","Monday","Tuesday","Wednessday","Thursday","Friday","Saturday"]
plt.xticks(count_weekday_casual.index,labels)
plt.show()
```

# -> The maximum number of rides for casual riders was on Sunday and lowest was recorded on Monday.

# -> The maximum number of rides for annual members was on Thursday and lowest was recorded on Tuesday.

# Calculating the average ride length for users, members and casual riders by weekday.

```python
# Calculating the average ride length for users  by weekday.
user_avg_ridelen = clean_file.groupby("week_day")["ride_length"].mean(numeric_only = False)
user_avg_ridelen
```

```python
# Calculating the average ride length for members  by weekday.
member_avg_ridelen = clean_file[clean_file["member_casual"]=="member"].groupby("week_day")["ride_length"].mean(numeric_only=False)
member_avg_ridelen
```

```python
# Calculating the average ride length for casual by weekday.
casual_avg_ridelen = clean_file[clean_file["member_casual"]=="casual"].groupby("week_day")["ride_length"].mean(numeric_only=False)
casual_avg_ridelen
```

# To plot weekday vs aveage ride length

```python
plt.figure(figsize =(10,8))
plt.plot(user_avg_ridelen/pd.Timedelta(minutes=1))
plt.plot(member_avg_ridelen/pd.Timedelta(minutes=1))
plt.plot(casual_avg_ridelen/pd.Timedelta(minutes=1))
plt.title("The average ride length for different users by weekday")
plt.legend(["user","member","casual"])
labels = ["Sunday","Monday","Tuesday","Wednessday","Thursday","Friday","Saturday"]
plt.xticks(user_avg_ridelen.index,labels)
plt.show()
```

# -> The highest average ride length for casual riders is observed on Tuesday which is approximately 50 minutes.
